```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
plt.rcParams.update({'font.family':'Nimbus Roman'})
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

```python
train = pd.read_csv("archive/KDDTrain+.txt", header=None)
```

```python
dos_attacks = ['apache2','back','land','neptune','mailbomb','pod','processtable','smurf','teardrop','udpstorm','worm']
probe_attacks = ['ipsweep','mscan','nmap','portsweep','saint','satan']
privilege_attacks = ['buffer_overflow','loadmdoule','perl','ps','rootkit','sqlattack','xterm']
access_attacks = ['ftp_write','guess_passwd','http_tunnel','imap','multihop','named','phf','sendmail','snmpgetattack','snmpguess','spy','warezclient','warezmaster','xclock','xsnoop']
attack_labels = ['Normal','DoS','Probe','Privilege','Access']

def mapper(a):
    if a in dos_attacks:
        return 1
    if a in probe_attacks:
        return 2
    if a in privilege_attacks:
        return 3
    if a in access_attacks:
        return 4
    return 0

train[41] = train[41].apply(mapper)
```

```python
fig = plt.figure(figsize=(28, 24), dpi =300)
c=0
for name in train.columns:
    if name==41:
        continue
    c+=1
    ax = fig.add_subplot(7, 6, c)
    ax.hist(train[name])
    ax.set_title(name)
    if train[name].dtype == np.dtype(object):
        ax.tick_params(axis='x', labelrotation=90)
plt.xticks(ha='right')
plt.tight_layout()
fig.savefig('resources/distribution.eps')
plt.show()
```

```python
fig = plt.figure(dpi=300, figsize=(6, 3))
ax = fig.add_subplot()
ax.barh(np.arange(5), [train[41][train[41]==i].size for i in range(5)], height=0.5, align='center', zorder=2, color="#4B9968")
for i in range(5):
    ax.annotate("%s"%train[41][train[41]==i].size, xy=(train[41][train[41]==i].size+500, i), textcoords='data', va='center')
ax.xaxis.set_minor_locator(MultipleLocator(5000))
ax.set_yticks(np.arange(5))
ax.set_xlim(0, 80000)
ax.set_yticklabels(attack_labels)
ax.grid(axis='x', zorder=0, linewidth=0.3)
ax.set_xlabel("Number of Observations")
ax.set_ylabel("Attacks")
ax.set_title("Distribution of Attack Type in the Dataset")
plt.tight_layout()
plt.savefig("resources/distro.png", dpi=300)
plt.savefig("resources/distro.pdf", dpi=300)
plt.show()
```
